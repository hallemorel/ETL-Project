```python
import pandas as pd
```

```python
#Read in 2014 National Names 
#Correct Column Headers
#Add Rank Column
national_names = pd.read_csv("/Users/hallemorel/Downloads/NationalFinal_2014.csv",sep=',', header=None)
national_names = national_names.rename(columns={0:'Name',1:'Gender',2:'National_Count'})
national_names['National_Rank']= national_names['National_Count'].rank(ascending=0)
national_names.head()
```

```python
#Read in NYC Names
#Adjust Column Headers
#Adjust Gender Column to match other tables
nyc_names = pd.read_csv("/Users/hallemorel/Downloads/Final_NYC.csv")
nyc_names = nyc_names.rename(columns={'brth_yr':'Year','gndr':'Gender','nm':'Name','Count':'City_Count','Rank':'City_Rank'})
nyc_names = nyc_names.replace({'FEMALE': 'F', 'MALE':'M'})
nyc_names.drop('Year', axis=1, inplace=True)
nyc_names.head()
```

```python
#Read in State CSV 
#Add in Rank Column
ny_state= pd.read_csv("/Users/hallemorel/Downloads/state_2014final.csv")
ny_state['State_Rank']= ny_state['State_Count'].rank(ascending=0)
ny_state.drop('Year', axis=1, inplace=True)
ny_state.head()
```

```python
#Combine National and NYC Data Frames
new_df=pd.merge(national_names,nyc_names, on=['Name','Gender'])
new_df
```

```python
#Combine above Data Frame with State Data Frame
final_df = pd.merge(new_df, ny_state, on=['Name','Gender'])
final_df.head()
#Save To CSV
final_df.to_csv('FINALNAME_2014.csv')                        
```
